#  <font color="blue"> 1️⃣7️⃣  **CHECK OUTLIERS**  </font>


```python
plt.figure(figsize=(8,4))

sns.boxplot(x=df["unit_sales"])

plt.title("Outlier Detection in Unit Sales")

plt.show()

show_interpretation(
    "Unit Sales Outlier Detection",
    "A box plot is used to visualize the distribution of `unit_sales` and identify any potential outliers. This helps in understanding the spread of sales data and extreme values.",
    "#e0e7ff"
)
```

```python
# Calculate Outliers
Q1 = df["unit_sales"].quantile(0.25)
Q3 = df["unit_sales"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[
    (df["unit_sales"] < lower_bound) |
    (df["unit_sales"] > upper_bound)
]

print("Number of outliers:", len(outliers))

show_interpretation(
    "Unit Sales Number Outlier",
    "Numnber of outliers:- 55",
    "#e0e7ff"
)
```

```python
# Showing Outliers using Scatter plot
plt.figure(figsize=(12,5))

plt.scatter(
    df.index,
    df["unit_sales"],
    alpha=0.6,
    label="Normal Data"
)

plt.scatter(
    outliers.index,
    outliers["unit_sales"],
    color="red",
    label="Outliers"
)

plt.title("Unit Sales with Outliers Highlighted")
plt.xlabel("Date")
plt.ylabel("Unit Sales")

plt.legend()

plt.show()
```

#  <font color="blue"> 1️⃣8️⃣ **HANDLE OUTLIERS** </font>


```python
cap_value = df["unit_sales"].quantile(0.99)

df["unit_sales_capped"] = np.where(
    df["unit_sales"] > cap_value,
    cap_value,
    df["unit_sales"]
)

show_interpretation(
    "Outlier Treatment",
    "Extreme sales values are capped at the 99th percentile to reduce their impact and improve model stability.",
    "#e0e7ff"
)

df.head()
```

```python
def adf_test(series):

    result = adfuller(series)

    print("ADF Statistic:", result[0])
    print("P-value:", result[1])

    if result[1] <= 0.05:
        print("Result: Data is Stationary")
    else:
        print("Result: Data is Not Stationary")

adf_test(df["unit_sales_capped"])

show_interpretation(
    "ADF Test for Stationarity",
    "The Augmented Dickey-Fuller (ADF) test is performed on the capped unit sales data to check for stationarity. A low p-value (typically <= 0.05) indicates that the data is stationary, a crucial assumption for many time series models.",
    "#e0e7ff"
)
```

# <font color="blue"> 3️⃣4️⃣ **MODEL EVALUATION** </font>


```python
def evaluate_model(actual, predicted, model_name):

    mae = mean_absolute_error(actual, predicted)

    rmse = np.sqrt(mean_squared_error(actual, predicted))

    mape = np.mean(
        np.abs((actual - predicted) / np.where(actual == 0, 1, actual))
    ) * 100

    print(model_name)
    print("MAE:", mae)
    print("RMSE:", rmse)
    print("MAPE:", mape)
    print("-" * 40)

evaluate_model(test, arima_prediction, "ARIMA Model")

evaluate_model(test, sarima_prediction, "SARIMA Model")

show_interpretation(
    "ARIMA and SARIMA Model Evaluation Metrics",
    "This section evaluates both the ARIMA and SARIMA models using key metrics: Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), and Mean Absolute Percentage Error (MAPE). These metrics quantify the accuracy of each model's predictions against actual sales data.",
    "#e0e7ff"
)
```

```python
#@title UI
%%writefile app.py

import streamlit as st
import joblib
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# PAGE CONFIGURATION
# ==============================
st.set_page_config(
    page_title="Clothing Store Sales Forecasting App",
    page_icon="📊",
    layout="wide"
)

# ==============================
# CUSTOM CSS - SAME BLUE + YELLOW 3D UI
# ==============================
css = """
<style>

[data-testid="stAppViewContainer"] {
background-image:
linear-gradient(120deg, rgba(2,6,23,0.82), rgba(30,64,175,0.72), rgba(250,204,21,0.22)),
url("https://images.unsplash.com/photo-1551288049-bebda4e38f71");
background-size: cover;
background-position: center;
background-repeat: no-repeat;
background-attachment: fixed;
}

[data-testid="stHeader"] {
background: rgba(0,0,0,0);
}

.block-container {
background: rgba(255,255,255,0.93);
padding: 2.2rem;
border-radius: 28px;
box-shadow:
0px 18px 40px rgba(0,0,0,0.35),
inset 0px 0px 18px rgba(255,255,255,0.35);
border: 2px solid rgba(250,204,21,0.55);
}

.main-title {
text-align: center;
font-size: 50px;
font-weight: 900;
color: #0f172a;
text-shadow: 3px 3px 0px #facc15, 6px 6px 15px rgba(37,99,235,0.35);
letter-spacing: 1px;
}

.sub-title {
text-align: center;
font-size: 21px;
font-weight: 600;
color: #1e3a8a;
margin-bottom: 25px;
}

.card {
background: linear-gradient(145deg, #facc15, #2563eb);
color: white;
padding: 24px;
border-radius: 22px;
text-align: center;
font-size: 21px;
font-weight: 800;
box-shadow:
8px 8px 18px rgba(15,23,42,0.35),
-5px -5px 14px rgba(255,255,255,0.75);
border: 2px solid rgba(255,255,255,0.65);
transition: all 0.3s ease-in-out;
}

.card:hover {
transform: translateY(-8px) scale(1.03);
box-shadow:
12px 14px 26px rgba(15,23,42,0.45),
0px 0px 25px rgba(250,204,21,0.75);
}

h1, h2, h3 {
color: #1e3a8a;
font-weight: 800;
}

.stButton > button {
background: linear-gradient(135deg, #facc15, #f97316);
color: #0f172a;
font-size: 18px;
font-weight: 800;
border-radius: 15px;
padding: 0.7rem 1.4rem;
border: none;
box-shadow: 0px 8px 18px rgba(249,115,22,0.45);
transition: all 0.25s ease-in-out;
}

.stButton > button:hover {
transform: translateY(-4px) scale(1.02);
background: linear-gradient(135deg, #2563eb, #1e40af);
color: white;
box-shadow: 0px 10px 25px rgba(37,99,235,0.55);
}

.stDownloadButton > button {
background: linear-gradient(135deg, #2563eb, #1d4ed8);
color: white;
font-size: 17px;
font-weight: 800;
border-radius: 15px;
padding: 0.7rem 1.4rem;
border: none;
box-shadow: 0px 8px 18px rgba(37,99,235,0.45);
}

.stDownloadButton > button:hover {
background: linear-gradient(135deg, #facc15, #f97316);
color: #0f172a;
transform: translateY(-4px);
}

[data-testid="stSidebar"] {
background: linear-gradient(180deg, #0f172a, #1e3a8a, #facc15);
color: white;
}

[data-testid="stSidebar"] * {
color: white;
}

[data-testid="stMetric"] {
background: linear-gradient(145deg, #eff6ff, #fef9c3);
padding: 18px;
border-radius: 18px;
box-shadow: 6px 6px 16px rgba(15,23,42,0.20);
border-left: 6px solid #facc15;
}

[data-testid="stFileUploader"] {
background: linear-gradient(145deg, #eff6ff, #fefce8);
padding: 16px;
border-radius: 18px;
border: 2px dashed #2563eb;
}

.footer {
text-align: center;
color: #1e3a8a;
font-weight: 900;
font-size: 19px;
text-shadow: 1px 1px 0px #facc15;
}

</style>
"""

st.markdown(css, unsafe_allow_html=True)

# ==============================
# LOAD MODEL
# ==============================
model = joblib.load("best_timeseries_model.pkl")
best_model_name = joblib.load("best_model_name.pkl")

# ==============================
# SIDEBAR
# ==============================
st.sidebar.title("📌 Project Menu")
st.sidebar.success(f"Best Model: {best_model_name}")

st.sidebar.header("Business Objective")
st.sidebar.write("Predict daily demand for one product-store combination.")
st.sidebar.write("Reduce stockouts, overstocking, and inventory cost.")
st.sidebar.write("Improve stock replenishment decisions.")

st.sidebar.header("Selected Scope")
st.sidebar.write("Clothing Store")
st.sidebar.write("T-Shirt")
st.sidebar.write("Target Variable: unit_sales")

st.sidebar.header("How to Use")
st.sidebar.write("1. Select future days")
st.sidebar.write("2. Choose chart type")
st.sidebar.write("3. Click Generate Forecast")
st.sidebar.write("4. Download forecast result")

st.sidebar.header("Model Meaning")
st.sidebar.write("ARIMA → Non-seasonal forecasting")
st.sidebar.write("SARIMA → Seasonal forecasting")
st.sidebar.write("RMSE → Model selection metric")

# ==============================
# MAIN TITLE
# ==============================
st.markdown("<div class='main-title'>🚀 Retail Demand Forecasting Dashboard 📊</div>", unsafe_allow_html=True)
st.markdown("<div class='sub-title'>Clothing Store | T-Shirt | Daily Sales Forecasting using ARIMA / SARIMA</div>", unsafe_allow_html=True)

st.write("---")

# ==============================
# PROJECT INTRODUCTION
# ==============================
st.info("""
This dashboard is built for a retail demand forecasting project.
It focuses on predicting daily sales demand for Clothing Store and T-Shirt.
The forecast helps reduce overstocking, avoid stockouts, and improve inventory planning.
""")

# ==============================
# TOP CARDS
# ==============================
c1, c2, c3, c4 = st.columns(4)

with c1:
    st.markdown("<div class='card'>🏬 Clothing Store</div>", unsafe_allow_html=True)

with c2:
    st.markdown("<div class='card'>📦 T-Shirt</div>", unsafe_allow_html=True)

with c3:
    st.markdown(f"<div class='card'>✅ Best Model<br>{best_model_name}</div>", unsafe_allow_html=True)

with c4:
    st.markdown("<div class='card'>📉 Metric<br>Lowest RMSE</div>", unsafe_allow_html=True)

st.write("---")

# ==============================
# FORECAST SETTINGS
# ==============================
st.header("🔮 Generate Future Demand Forecast")
st.info("This section predicts future daily unit sales. Example: 30 means next 30 days demand forecast.")

col1, col2, col3 = st.columns(3)

with col1:
    steps = st.slider(
        "Select number of future days",
        min_value=1,
        max_value=365,
        value=30
    )

with col2:
    chart_type = st.selectbox(
        "Choose chart type",
        ["Line Chart", "Area Chart", "Bar Chart"]
    )

with col3:
    show_table = st.radio(
        "Show forecast table?",
        ["Yes", "No"],
        horizontal=True
    )

# ==============================
# FORECAST BUTTON
# ==============================
if st.button("🚀 Generate Demand Forecast"):

    forecast = model.forecast(steps=steps)

    forecast_df = pd.DataFrame({
        "Future_Day": range(1, steps + 1),
        "Forecast_Unit_Sales": forecast
    })

    st.success("Demand forecast generated successfully!")

    st.subheader("📌 Forecast Summary")

    first_value = forecast_df["Forecast_Unit_Sales"].iloc[0]
    last_value = forecast_df["Forecast_Unit_Sales"].iloc[-1]

    if last_value > first_value:
        trend = "Increasing Demand 📈"
    elif last_value < first_value:
        trend = "Decreasing Demand 📉"
    else:
        trend = "Stable Demand ➖"

    s1, s2, s3, s4 = st.columns(4)

    with s1:
        st.metric("Future Days", steps)

    with s2:
        st.metric("Average Demand", round(forecast_df["Forecast_Unit_Sales"].mean(), 2))

    with s3:
        st.metric("Highest Demand", round(forecast_df["Forecast_Unit_Sales"].max(), 2))

    with s4:
        st.metric("Demand Trend", trend)

    if show_table == "Yes":
        st.subheader("📋 Forecast Table")
        st.dataframe(forecast_df, use_container_width=True)

    st.subheader("📈 Forecast Visualization")

    chart_data = forecast_df.set_index("Future_Day")

    if chart_type == "Line Chart":
        st.line_chart(chart_data)

    elif chart_type == "Area Chart":
        st.area_chart(chart_data)

    else:
        st.bar_chart(chart_data)

    st.subheader("🔍 Forecast Confidence Interval")

    try:
        forecast_result = model.get_forecast(steps=steps)
        confidence_interval = forecast_result.conf_int()
        confidence_interval = confidence_interval.reset_index(drop=True)
        confidence_interval["Future_Day"] = range(1, steps + 1)

        st.dataframe(confidence_interval, use_container_width=True)

    except:
        st.warning("Confidence interval is not available for this saved model.")

    st.subheader("📊 Detailed Forecast Graph")

    fig, ax = plt.subplots(figsize=(12, 5))
    ax.plot(
        forecast_df["Future_Day"],
        forecast_df["Forecast_Unit_Sales"],
        marker="o",
        linewidth=2
    )

    ax.set_title(f"{best_model_name} Future Demand Forecast")
    ax.set_xlabel("Future Days")
    ax.set_ylabel("Forecast Unit Sales")
    ax.grid(True)

    st.pyplot(fig)

    csv = forecast_df.to_csv(index=False).encode("utf-8")

    st.download_button(
        label="⬇️ Download Demand Forecast CSV",
        data=csv,
        file_name="retail_demand_forecast.csv",
        mime="text/csv"
    )

# ==============================
# DATASET UPLOAD
# ==============================
st.write("---")
st.header("📂 Upload Dataset for Analysis")

st.info("""
Upload train_2016.csv for latest sales data preview.
You may also upload oil.csv or transactions.csv only for viewing and analysis.
This upload section is not used for retraining the saved model.
""")

uploaded_file = st.file_uploader(
    "Upload CSV file",
    type=["csv"]
)

if uploaded_file is not None:

    df = pd.read_csv(uploaded_file)

    st.success("Dataset uploaded successfully!")

    st.subheader("👀 Dataset Preview")
    st.dataframe(df.head(), use_container_width=True)

    st.subheader("📊 Dataset Overview")

    d1, d2, d3, d4 = st.columns(4)

    with d1:
        st.metric("Rows", df.shape[0])

    with d2:
        st.metric("Columns", df.shape[1])

    with d3:
        st.metric("Missing Values", df.isnull().sum().sum())

    with d4:
        st.metric("Duplicate Rows", df.duplicated().sum())

    st.subheader("📌 Column Names")
    st.write(df.columns.tolist())

    numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

    if len(numeric_cols) > 0:

        selected_col = st.selectbox(
            "Select numeric column for analysis",
            numeric_cols
        )

        st.subheader(f"📈 Trend of {selected_col}")
        st.line_chart(df[selected_col])

        st.subheader("📌 Basic Statistics")
        st.dataframe(df[selected_col].describe())

        st.subheader("📉 Missing Values by Column")

        missing_df = pd.DataFrame({
            "Column": df.columns,
            "Missing_Count": df.isnull().sum().values
        })

        st.dataframe(missing_df, use_container_width=True)

    else:
        st.warning("No numeric column found.")

# ==============================
# PROJECT METHODOLOGY
# ==============================
st.write("---")
st.header("🧪 Project Methodology")

st.write("""
1. Imported sales data from 2013 to 2016.
2. Combined yearly sales data into one dataset.
3. Filtered data for Clothing Store and T-Shirt.
4. Added missing dates to maintain daily time series continuity.
5. Merged oil price and transaction data as external features.
6. Handled missing values and capped outliers.
7. Checked stationarity using transformation and differencing.
8. Built ARIMA and SARIMA models.
9. Evaluated models using MAE, MAPE, and RMSE.
10. Selected the best model using the lowest RMSE.
""")

# ==============================
# MODEL EXPLANATION
# ==============================
st.write("---")
st.header("🧠 Project Interpretation")

st.info("""
This dashboard forecasts future daily demand for a selected retail product-store combination.

ARIMA is used to capture non-seasonal demand patterns.

SARIMA is used to capture seasonal demand patterns.

Oil prices and transactions were considered as external data during feature engineering and analysis.

The best model is selected using RMSE. Lower RMSE means the model prediction error is lower.
""")

# ==============================
# FOOTER
# ==============================
st.write("---")
st.markdown(
    "<div class='footer'>🚀 Retail Demand Forecasting Project | Clothing Store | T-Shirt | ARIMA vs SARIMA</div>",
    unsafe_allow_html=True
)
```
